```python
# imports

import os 
from dotenv import load_dotenv
from openai import OpenAI
import requests
from IPython.display import display, update_display, Markdown
```

```python
# constants
base_url = "https://generativelanguage.googleapis.com/v1beta/openai/"

MODEL_GEMINI = 'gemini-2.5-flash-lite'
MODEL_GPT = 'gpt-4o-mini'
MODEL_LLAMA = 'llama3.2'
```

```python
# here is the question; type over this to ask something new

question = """
Please explain what this code does and why:
yield from {book.get("author") for book in books if book.get("author")}
"""

system_prompt = """
you are an expert technical tutor, which reads and analyzes the question clearly that students ask, and provides a step by step 
easy to understand explanation with few good examples as well if applicable, also you may ask follow up topics to the students
Respond in markdown without code blocks.
"
"""
```

```python
# Get gpt-4o-mini to answer, with streaming

def ask_question(user_question):
    stream = openai.chat.completions.create(
        model = MODEL_GEMINI,
        messages=[{
            'role': 'system', 'content': system_prompt,
            'role': 'user', 'content': user_question
        }],
        stream=True
    )

    response = ''
    display_handle = display(Markdown(''), display_id = True)

    for chunk in stream:
        response += chunk.choices[0].delta.content or ''
        update_display(Markdown(response), display_id=display_handle.display_id)
```

```python
ask_question(question)
```
