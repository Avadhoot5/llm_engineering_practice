```python
# importing the libraries

import os
from dotenv import load_dotenv
from openai import OpenAI
from IPython.display import display, Markdown
import gradio as gr
```

```python
## Load api keys

load_dotenv(override=True)
gemini_api_key = os.getenv('GOOGLE_API_KEY')
open_router_api_key = os.getenv('OPENROUTER_API_KEY')

if open_router_api_key and open_router_api_key.startswith('sk-or'):
    print('Open Router API key found and good so far')
else:
    print('Api Key not found')

if gemini_api_key and gemini_api_key.startswith('AIz'):
    print('API key found and looks good so far!')
else:
    print('API key not found, please check')
```

```python
gemini_url = "https://generativelanguage.googleapis.com/v1beta/openai/"
open_router_url = "https://openrouter.ai/api/v1"

gemini = OpenAI(base_url=gemini_url, api_key=gemini_api_key)
open_router = OpenAI(api_key=open_router_api_key, base_url=open_router_url)

gemini_2 = 'gemini-2.5-flash-lite'
gemini_3 = 'gemini-3.1-flash-lite'

# Open router free models 

gemini_model_gemma_4_31b = 'google/gemma-4-31b-it:free'
gpt_oss_model = 'openai/gpt-oss-120b:free'
inclusionai_model_ring_2_1t = 'inclusionai/ring-2.6-1t:free'
baidu_model = "baidu/cobuddy:free"
deepseek_v4_flash_model = "deepseek/deepseek-v4-flash:free"
```

```python
system_message = '''
You are an helpful assistant for an Airline called FlightyAI.
Give short, courteous answers, no more than 1 sentence.
Always be accurate. If you don't know the answer, say so.
'''
```

```python
def chat(message, history):
    history = [{'role': h['role'], 'role': h['content']} for h in history]
    messages = [{'role': 'system', 'content': system_message}] + history + [{'role': 'user', 'content': message}]
    response = open_router.chat.completions.create(model=gpt_oss_model, messages=messages)
    return response.choices[0].message.content

app = gr.ChatInterface(
    fn=chat,
    title='FlightyAI Chat Support'
)

app.launch()
```

## Tools

Tools are an incredibly powerful feature provided by the frontier LLMs.

With tools, you can write a function, and have the LLM call that function as part of its response.

Sounds almost spooky.. we're giving it the power to run code on our machine?

Well, kinda.

```python
# Let's start by making a useful function

ticket_prices = {'mumbai': 5000, 'delhi': 3000, 'tokyo': 50000, 'banglore': 6000}

def get_ticket_price(destination_city):
    print(f'Tool called for city {destination_city}')
    ticket_price = ticket_prices.get(destination_city.lower(), "Unknown ticket price")
    return f'The ticket price for the {destination_city} is {ticket_price}'
```
