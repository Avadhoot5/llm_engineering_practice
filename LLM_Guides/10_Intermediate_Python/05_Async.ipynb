### Exercise 2 – Parallel Coroutines

```python
async def fetch_data():
    await asyncio.sleep(1)
    return 'Fetched the data'

async def process_data():
    await asyncio.sleep(1)
    return 'Processed the data'

async def save_data():
    await asyncio.sleep(1)
    return 'Saved the data'

async def main_data():
    results = await asyncio.gather(
        fetch_data(),
        process_data(),
        save_data()
    )
    print(results)

await main_data()
```

### Exercise 4 – Compare Blocking vs Async

```python
import time

def blocking():
    time.sleep(1)
    print("Blocking done")


start_time = time.time()

for _ in range(3):
    blocking()

end_time = time.time()

total_time = round(end_time - start_time, 2)
print(f'Function blocking time {total_time}')

async def non_blocking():
    await asyncio.sleep(1)
    print("Async done")

start_time = time.time()

tasks = [non_blocking() for _ in range(3)]
await asyncio.gather(*tasks)
end_time = time.time()

total_time = round(end_time - start_time, 2)
print(f'Function non-blocking time {total_time}')
```
